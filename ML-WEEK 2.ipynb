```python
#1. Print 1 to 5 using (for,while,do-while ) loops. 

# Using for loop
for i in range(1, 6):
    print(i, end=" ")
    
print("\n")

# Using while loop
x = 1
while x <= 5:
    print(x, end=" ")
    x += 1

print("\n")
    

# Using do-while loop simulation
x = 1
while True:
    print(x, end=" ")
    x += 1
    if x > 5:
        break
```

```python
#2. Print n  numbers of sum using (for,while,do-while ) loops. 

# Using for loop
n = 5
sum_result_for = 0

for i in range(1, n + 1):
    sum_result_for += i

print(f"The sum of the first {n} numbers using for loop is: {sum_result_for}")

# Using while loop
n = 5
sum_result_while = 0
i = 1

while i <= n:
    sum_result_while += i
    i += 1

print(f"The sum of the first {n} numbers using while loop is: {sum_result_while}")

# Using do-while loop simulation
n = 5
sum_result_do_while = 0
i = 1

while True:
    sum_result_do_while += i
    i += 1
    if i > n:
        break

print(f"The sum of the first {n} numbers using do-while loop simulation is: {sum_result_do_while}")

```

```python
#3. Calculate Factorial using (for,while,do-while ) loops.
# Using for loop
n = 5
factorial_for = 1

for i in range(1, n + 1):
    factorial_for *= i

print(f"The factorial of {n} using for loop is: {factorial_for}")

# Using while loop
n = 5
factorial_while = 1
i = 1

while i <= n:
    factorial_while *= i
    i += 1

print(f"The factorial of {n} using while loop is: {factorial_while}")

# Using do-while loop simulation
n = 5
factorial_do_while = 1
i = 1

while True:
    factorial_do_while *= i
    i += 1
    if i > n:
        break

print(f"The factorial of {n} using do-while loop simulation is: {factorial_do_while}")
```

```python
#4. Count no. OF vowels in string using loop. 
# Using for loop to count vowels in a string
input_string = "Hello, World!"
vowel_count = 0

for char in input_string:
    # Check if the character is a vowel (case-insensitive)
    if char.lower() in "aeiou":
        vowel_count += 1

print(f"The number of vowels in the string is: {vowel_count}")
```

```python
#5. Print any 3 standard pattern using loops.

# Right Triangle Pattern
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    print()
    
    
print("\n")
# Square Pattern
rows = 5

for i in range(1, rows + 1):
    for j in range(1, rows + 1):
        print("*", end=" ")
    print()
    
print("\n")  
# Pyramid Pattern
rows = 5

for i in range(1, rows + 1):
    for j in range(1, rows - i + 1):
        print(" ", end=" ")
    for k in range(1, 2 * i):
        print("*", end=" ")
    print()
```
